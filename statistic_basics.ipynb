Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
def filter_purch(sum_purchases, visits, mean_purch, purch_std):
    purch_clear = list()
    visits_clear = list()
    for i in range(len(sum_purchases)):
        if sum_purchases[i] <= (mean_purch + 2 * purch_std):
            purch_clear.append(sum_purchases[i])
            visits_clear.append(visits[i])
    return purch_clear, visits_clear

purch_clear, visits_clear =  filter_purch(sum_purchases, visits, mean_purch, purch_std)
print(purch_clear)
print(visits_clear)
kkP3 = np.corrcoef(purch_clear, visits_clear)
kkP3
```
